Ejercicio 1


```python
# Ejercicio 1: Calculadora de Factura con IVA (Python)

IVA_TASA = 0.16  # La tasa de IVA en México (16%)

def calcular_factura():
    """
    Solicita los datos de un producto, calcula el subtotal, IVA y total,
    y muestra la factura completa en la salida estándar.
    """
    print("\n--- Calculadora de Factura con IVA ---")

    # 1. Función que pida: nombre del producto, precio unitario, cantidad de venta
    try:
        nombre_producto = input("Ingrese el nombre del producto: ")
        # Se asegura de que el precio y la cantidad sean números positivos.
        precio_unitario = float(input(f"Ingrese el precio unitario de {nombre_producto} (MXN): "))
        cantidad_venta = int(input(f"Ingrese la cantidad vendida de {nombre_producto}: "))

        if precio_unitario <= 0 or cantidad_venta <= 0:
            print("Error: El precio y la cantidad deben ser valores positivos.")
            return

    except ValueError:
        print("Error: Ingrese valores numéricos válidos para precio y cantidad.")
        return

    # 2. Calcular: subtotal (precio × cantidad), IVA (subtotal × 0.16), total (Subtotal + IVA)

    # Cálculo del subtotal
    subtotal = precio_unitario * cantidad_venta

    # Cálculo del IVA
    iva = subtotal * IVA_TASA

    # Cálculo del total
    total = subtotal + iva

    # 3. Mostrar en salida estándar la factura completa

    print("\n--- FACTURA COMPLETA ---")
    print(f"Producto: {nombre_producto}")
    print(f"Precio Unitario: ${precio_unitario:,.2f} MXN")
    print(f"Cantidad: {cantidad_venta}")
    print("-" * 30)
    print(f"Subtotal (sin IVA): ${subtotal:,.2f} MXN")
    print(f"IVA ({int(IVA_TASA * 100)}%): ${iva:,.2f} MXN")
    print("-" * 30)
    print(f"TOTAL A PAGAR: ${total:,.2f} MXN")
    print("-" * 30)

# Ejecución del ejercicio 1
calcular_factura()
```

```python
# Ejercicio 2: Calculadora de Dimensiones para Cableado (Python)

def calcular_largo_fibra():
    """
    Solicita el largo, ancho de una habitación y la cantidad de fibras a instalar.
    Calcula el total de metros de fibra óptica necesaria.
    """
    print("\n--- Calculadora de Largo de Instalación de Fibra Óptica ---")

    # Pedir dimensiones: largo y ancho de habitación, cantidad de fibras a instalar
    try:
        largo = float(input("Ingrese el largo de la habitación (en metros): "))
        ancho = float(input("Ingrese el ancho de la habitación (en metros): "))

        # Se asume que la instalación sigue el perímetro (el escenario más común)
        cantidad_fibras = int(input("Ingrese la cantidad de fibras ópticas a instalar (ej. 2): "))

        if largo <= 0 or ancho <= 0 or cantidad_fibras <= 0:
            print("Error: Todas las dimensiones y la cantidad deben ser valores positivos.")
            return

    except ValueError:
        print("Error: Ingrese valores numéricos válidos para dimensiones y cantidad.")
        return

    # Calcular perímetro total de fibra necesaria x la cantidad de fibras a instalar

    # 1. Calcular el perímetro de la habitación (Largo + Ancho) * 2
    perimetro_habitacion = (largo + ancho) * 2

    # 2. Calcular el largo total de fibra óptica
    largo_total_fibra = perimetro_habitacion * cantidad_fibras

    # Mostrar en salida estándar cuántos metros de fibra necesito

    print("\n--- Resultado del Cálculo ---")
    print(f"Dimensiones de la habitación: {largo}m x {ancho}m")
    print(f"Perímetro de la instalación: {perimetro_habitacion:.2f} metros")
    print(f"Cantidad de fibras a instalar: {cantidad_fibras}")
    print("-" * 30)
    print(f"Metros de fibra óptica TOTALES necesarios: {largo_total_fibra:.2f} metros")
    print("-" * 30)

# Ejecución del ejercicio 2
calcular_largo_fibra()
```

```python
# Ejercicio 3: Sistema de Ventas utilizando Pandas (Python)

import pandas as pd
from datetime import datetime

# Definición del menú predeterminado y sus precios base
MENU_PRODUCTOS = {
    "pan": 15.00,
    "refresco": 22.50,
    "taco": 18.00,
    "pizza": 35.00
}

def registrar_venta(ventas_df):
    """
    Captura los datos de una venta, calcula el subtotal y lo añade al DataFrame.
    Utiliza match case para determinar el precio unitario de los productos predefinidos.
    """
    print("\n--- Registrar Nueva Venta ---")

    # Capturar: producto, cantidad
    producto = input("Ingrese el nombre del producto (Pan, Refresco, Taco, Pizza): ").lower().strip()

    try:
        cantidad = int(input(f"Ingrese la cantidad vendida de {producto}: "))
        if cantidad <= 0:
            print("Error: La cantidad debe ser positiva.")
            return ventas_df
    except ValueError:
        print("Error: Ingrese una cantidad numérica válida.")
        return ventas_df

    precio = 0

    # Usar match case para determinar el precio (ejemplo de menú predeterminado)
    match producto:
        case "pan":
            precio = MENU_PRODUCTOS["pan"]
        case "refresco":
            precio = MENU_PRODUCTOS["refresco"]
        case "taco":
            precio = MENU_PRODUCTOS["taco"]
        case "pizza":
            precio = MENU_PRODUCTOS["pizza"]
        case _:
            # Opción para productos no listados
            try:
                precio = float(input(f"Producto no listado. Ingrese precio unitario para {producto}: "))
                if precio <= 0:
                    print("Error: El precio debe ser positivo.")
                    return ventas_df
            except ValueError:
                print("Error: Ingrese un precio numérico válido.")
                return ventas_df

    # Calcular el subtotal
    subtotal = precio * cantidad

    # Preparar el registro para el DataFrame
    nueva_venta = pd.DataFrame([{
        'Venta_ID': ventas_df['Venta_ID'].max() + 1 if not ventas_df.empty else 1,
        'Producto': producto.capitalize(),
        'Cantidad': cantidad,
        'Precio_Unitario': precio,
        'Subtotal': subtotal,
        'Hora': datetime.now().strftime("%H:%M:%S")
    }])

    # Añadir la nueva venta al DataFrame
    ventas_df = pd.concat([ventas_df, nueva_venta], ignore_index=True)
    print(f"✅ Venta de {cantidad} x {producto.capitalize()} registrada con éxito.")
    return ventas_df

def generar_corte_del_dia(ventas_df):
    """
    Genera el corte del día mostrando el detalle de ventas y el total general.
    """
    if ventas_df.empty:
        print("\n--- CORTE DEL DÍA ---")
        print("No se han registrado ventas hoy.")
        return

    # Calcular el total de ventas
    total_ventas = ventas_df['Subtotal'].sum()

    print("\n--- REPORTE DETALLADO DE VENTAS DEL DÍA ---")
    print(ventas_df.to_string(index=False)) # Muestra todo el DataFrame sin el índice
    print("=" * 55) # Línea de separación mejorada
    print(f"Total de ventas registradas: {len(ventas_df)}")
    # Se asegura de que el TOTAL se muestre con claridad
    print(f"TOTAL INGRESOS DEL DÍA: ${total_ventas:,.2f} MXN")
    print("=" * 55) # Línea de cierre

def menu_principal():
    """
    Menú interactivo para el sistema de ventas.
    """
    # Inicialización del DataFrame de ventas
    columnas = ['Venta_ID', 'Producto', 'Cantidad', 'Precio_Unitario', 'Subtotal', 'Hora']
    ventas_df = pd.DataFrame(columns=columnas)

    while True:
        print("\n==============================")
        print("  SISTEMA DE REGISTRO DIARIO ")
        print("==============================")
        print("1. Registrar Venta")
        print("2. Generar Corte del Día")
        print("3. Salir")

        opcion = input("Seleccione una opción: ")

        match opcion:
            case '1':
                ventas_df = registrar_venta(ventas_df)
            case '2':
                generar_corte_del_dia(ventas_df)
            case '3':
                print("Cerrando el sistema. ¡Hasta pronto!")
                # Mostrar el total de ventas antes de salir
                if not ventas_df.empty:
                    total_final = ventas_df['Subtotal'].sum()
                    print(f"\nResumen final: Total de ingresos del día: ${total_final:,.2f} MXN")
                break
            case _:
                print("Opción no válida. Intente de nuevo.")

# Ejecución del ejercicio 3
if __name__ == "__main__":
    # Se añade la verificación para asegurar que Pandas esté disponible
    try:
        menu_principal()
    except ImportError:
        print("\n--- ERROR DE DEPENDENCIA ---")
        print("La librería 'pandas' no está instalada.")
        print("Instale con: pip install pandas")
```

```python
# Ejercicio 4: Sistema de Calificación de Películas (Python)

# Lista de 5 películas iniciales para calificar (nombre: calificación)
# Se usa un diccionario para almacenar los datos (calificación inicial 0)
peliculas = {
    "El Padrino": 0,
    "Interestelar": 0,
    "El Viaje de Chihiro": 0,
    "La La Land": 0,
    "Parásitos": 0
}

def mostrar_menu():
    """Muestra las opciones disponibles al usuario."""
    print("\n===============================")
    print("  SISTEMA DE CALIFICACIÓN CINÉFILA")
    print("===============================")
    print("1. Calificar Película")
    print("2. Ver Recomendaciones (5+ Estrellas)")
    print("3. Ver todas las Calificaciones")
    print("4. Salir")
    opcion = input("Seleccione una opción: ")
    return opcion

def calificar_pelicula():
    """Permite al usuario calificar una película de la lista."""
    print("\n--- Calificar Película ---")

    # Muestra las películas disponibles
    nombres_peliculas = list(peliculas.keys())
    for i, nombre in enumerate(nombres_peliculas):
        print(f"{i + 1}. {nombre} (Calificación actual: {peliculas[nombre]} estrellas)")

    try:
        seleccion = int(input("Ingrese el número de la película a calificar: "))
        if 1 <= seleccion <= len(nombres_peliculas):
            pelicula_seleccionada = nombres_peliculas[seleccion - 1]

            calificacion = int(input("Ingrese su calificación (1 a 10 estrellas): "))

            # Filtro de validación de la calificación
            if 1 <= calificacion <= 10:
                peliculas[pelicula_seleccionada] = calificacion
                print(f"✅ ¡Gracias! '{pelicula_seleccionada}' ha sido calificada con {calificacion} estrellas.")
            else:
                print("Error: La calificación debe estar entre 1 y 10.")
        else:
            print("Error: Número de película no válido.")
    except ValueError:
        print("Error: Ingrese un número válido para la selección y la calificación.")

def ver_recomendaciones():
    """
    Filtro de recomendaciones: muestra solo películas con 5 o más estrellas.
    """
    print("\n--- RECOMENDACIONES (Calificación 5 o más) ---")

    recomendaciones_encontradas = False

    # Uso de la estructura 'if' para filtrar
    for pelicula, calificacion in peliculas.items():
        if calificacion >= 5:
            print(f"⭐ {pelicula} - Calificación: {calificacion} estrellas")
            recomendaciones_encontradas = True

    if not recomendaciones_encontradas:
        print("Aún no hay películas recomendadas con 5 o más estrellas. ¡A calificar!")

def ver_calificaciones():
    """Muestra todas las películas y sus calificaciones actuales."""
    print("\n--- TODAS LAS CALIFICACIONES ---")
    for pelicula, calificacion in peliculas.items():
        print(f"- {pelicula}: {calificacion} estrellas")

def sistema_peliculas():
    """Función principal del sistema de calificación de películas."""

    while True:
        opcion = mostrar_menu()

        # Uso de la estructura 'match case' para la navegación del menú
        match opcion:
            case '1':
                calificar_pelicula()
            case '2':
                ver_recomendaciones()
            case '3':
                ver_calificaciones()
            case '4':
                print("Saliendo del sistema de calificación. ¡Hasta luego!")
                break
            case _:
                print("Opción no válida. Por favor, ingrese un número del 1 al 4.")

# Ejecución del ejercicio 4
sistema_peliculas()
```
